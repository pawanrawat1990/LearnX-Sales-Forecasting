```python
import pandas as pd
import numpy as np
import zipfile
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#### Making a connection to zip file ###
conn=zipfile.ZipFile("D:/Python/LearnX Sales Forecasting/train_yhhx1Xs.zip")
```

```python
#### Importing CSV from zip ####
train=pd.read_csv(conn.open("train.csv"))
```

```python
test=pd.read_csv("D:/Python/LearnX Sales Forecasting/test_QkPvNLx.csv")
sample_submission=pd.read_csv("D:/Python/LearnX Sales Forecasting/sample_submission_pn2DrMq.csv")
```

```python
train["Temp"]="train"
test["Temp"]="test"
test["Sales"]=0
test["User_Traffic"]=0
```

```python
combi=pd.concat([train,test],axis=0,ignore_index=True)
```

```python
print(train.describe(percentiles=[0.01,0.05,0.95,0.99]).T)
combi.describe(percentiles=[0.01,0.05,0.95,0.99]).T
```

```python
print(combi["Course_Domain"].value_counts(dropna=False))
combi["Course_Type"].value_counts(dropna=False)
```

```python
combi.apply(lambda x : sum(x.isnull()))
```

```python
### As sales & User Traffic is only avaialble in training data, uisng outlier capping in train data only ###
numeric_col= train[["User_Traffic","Sales"]]
for i in numeric_col:
    train.loc[train[i]>train[i].quantile(0.99),i]=train[i].quantile(0.99)
    train.loc[train[i]<train[i].quantile(0.01),i]=train[i].quantile(0.01)
train.describe(percentiles=[0.01,0.05,0.95,0.99]).T
```

```python
combi_2=pd.concat([train,test],axis=0,ignore_index=True)
```

```python
### As Competition_Metric in both the tables , we can do outlier capping after combining these ####
numeric_col= combi_2[["Competition_Metric"]]
for i in numeric_col:
    combi_2.loc[combi_2[i]>combi_2[i].quantile(0.99),i]=combi_2[i].quantile(0.99)
```

```python
combi_2["Competition_Metric"] = combi_2.groupby(["Course_Domain","Course_Type"])["Competition_Metric"].transform(lambda x: x.fillna(x.mean()))
```

```python
import math
```

```python
combi_2["Year_Num"]=np.ceil(combi_2["Day_No"]/365)
```

```python
combi_2["Year_Day_Num"]=0
```

```python
combi_2.loc[combi_2["Year_Num"]==1,"Year_Day_Num"]=combi_2["Day_No"]
combi_2.loc[combi_2["Year_Num"]==2,"Year_Day_Num"]=combi_2["Day_No"]-365
combi_2.loc[combi_2["Year_Num"]==3,"Year_Day_Num"]=combi_2["Day_No"]-730
```

```python
combi_2["Months"]=np.ceil(combi_2["Year_Day_Num"]/30)
```

```python
combi_2.loc[combi_2["Months"]==13,"Months"]=12
```

```python
combi_2["Weeks"]=np.ceil(combi_2["Year_Day_Num"]/7)
```

```python
combi_2.loc[combi_2["Weeks"]==53,"Weeks"]=52
```

```python
cols=combi_2[["Year_Num","Months","Weeks","Long_Promotion","Public_Holiday","Short_Promotion"]]
for i in cols:
    combi_2[i]=combi_2[i].astype(object)
```

```python
dummy=pd.get_dummies(combi_2[["Course_Domain","Course_Type","Long_Promotion","Public_Holiday","Short_Promotion","Year_Num","Months","Weeks"]],drop_first=False)
combine=pd.concat([combi_2,dummy],axis=1)
```

```python
combine.drop(["Course_Domain","Course_Type","Long_Promotion","Public_Holiday","Short_Promotion","Year_Num","Months","Weeks","Day_No","ID","Course_ID","Year_Day_Num"],axis=1,inplace=True)
```

First predicting User Traffic for testing data ####

```python
training=combine.loc[combine["Temp"]=="train"]
testing=combine.loc[combine["Temp"]=="test"]
```

```python
training_1=training.drop(["Temp","Sales"],axis=1)
testing_1=testing.drop(["Temp","Sales","User_Traffic"],axis=1)
```

```python
x_train=training_1.drop(["User_Traffic"],1)
y_train=training_1["User_Traffic"]
```

```python
np.random.seed(324)
```

```python
from sklearn.model_selection import KFold, train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.30,random_state=43)
```

```python
from catboost import CatBoostRegressor
```

```python
cb_model = CatBoostRegressor(iterations=2000,
                             learning_rate=0.1,
                             depth=10,
                             eval_metric='RMSE',
                             loss_function='RMSE',
                             random_seed = 23,
                             bagging_temperature = 0.2,
                             od_type='Iter',
                             metric_period = 100,
                             od_wait=100)
```

```python
cb_model.fit(X_train, Y_train,
             eval_set=(X_test,Y_test),
             use_best_model=True,
             verbose=True)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rmse=sqrt(mean_squared_error(Y_test,cb_model.predict(X_test)))
```

```python
rmse
```

```python
testing_1["User_Traffic"]=cb_model.predict(testing_1)
```

```python
testing["User_Traffic"]=testing_1["User_Traffic"]
```
